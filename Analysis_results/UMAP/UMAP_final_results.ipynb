```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from sklearn import preprocessing
import umap.umap_ as umap
```

# **Diccionario de Colores**

```python
custom_colors = [(255/255,0/255,0/255), #red
               (0/255, 0/255, 0/255), #black
               (0/255, 0/255, 255/255), #blue
               (0/255, 144/255, 0/255), #green
               (255/255, 255/255, 126/255), #yellow
               (144/255, 144/255, 144/255), #gray
               (172/255, 135/255, 73/255), #brown
               (255/255, 192/255, 203/255), #pink
               (149/255, 53/255, 83/255), #purple
               (255/255, 165/255, 0/255) #orange
               ]
```

```python
print('custom_colors', type(custom_colors), 'custom colors element', type(custom_colors[0]))
my_cmap = LinearSegmentedColormap.from_list('UMAP_colors', custom_colors, N=10)
```

```python
color_layer1_df=pd.read_csv("data_pool1_pool_20211001-203037_Color.csv")
texture_layer1_df=pd.read_csv("data_pool1_pool_20211001-210021_Texture.csv")

color_layer2_df=pd.read_csv("data_conv2_block3_out_20211001-211701_Color.csv")
texture_layer2_df=pd.read_csv("data_conv2_block3_out_20211001-211038_Texture.csv")

color_layer3_df=pd.read_csv("data_conv3_block4_out_20211001-212717_Color.csv")
texture_layer3_df=pd.read_csv("data_conv3_block4_out_20211001-213213_Texture.csv")

color_layer4_df=pd.read_csv("data_conv4_block6_out_20211001-214906 Color.csv")
texture_layer4_df=pd.read_csv("data_conv4_block6_out_20211001-214150 Texture.csv")

color_layer5_df=pd.read_csv("data_conv5_block3_out_20211001-220255 Color.csv")
texture_layer5_df=pd.read_csv("data_conv5_block3_out_20211001-221216 Texture.csv")
```

# **Procesamiento - Visualización**

# Capa 1

```python
le = preprocessing.LabelEncoder()
feature_columns_header = ["f"+str(i) for i in range(1, color_layer1_df.shape[1])]
label_column_header = ['label']
color_layer1_df.columns = label_column_header + feature_columns_header
string_labels = ['red', 'black', 'blue', 'green', 'yellow', 'gray', 'brown', 'pink', 'purple', 'orange']
color_layer1_df['label'] = le.fit_transform(color_layer1_df['label'])
layer_data = color_layer1_df.iloc[:, 1:].values
reducer = umap.UMAP(random_state=42, n_neighbors=15, metric='euclidean')
embedding = reducer.fit_transform(layer_data)
test_colors = color_layer1_df.label
scatter_plot = plt.scatter(embedding[:, 0], embedding[:, 1], c=test_colors, cmap=my_cmap)
plt.gca().set_aspect('equal', 'datalim')
#title = 'UMAP projection - color dataset with ', dimensiones ,' dims'
plt.title('UMAP projection of the color dataset layer 1', fontsize=12)
handles, labels = scatter_plot.legend_elements()
labels = string_labels
plt.legend(handles, labels)
plt.show()
```

```python
feature_columns_header = ["f"+str(i) for i in range(1, texture_layer1_df.shape[1])]
label_column_header = ['label']
texture_layer1_df.columns = label_column_header + feature_columns_header
string_labels = ['squared', 'striped', 'flowers', 'leopard', 'polka', 'basic', 'paisley', 'agryle', 'crows feet', 'sequin']
texture_layer1_df['label'] = le.fit_transform(texture_layer1_df['label'])
layer_data = texture_layer1_df.iloc[:, 1:].values
reducer = umap.UMAP(random_state=42, n_neighbors=15, metric='euclidean')
embedding = reducer.fit_transform(layer_data)
test_colors = texture_layer1_df.label
scatter_plot = plt.scatter(embedding[:, 0], embedding[:, 1], c=test_colors, cmap=my_cmap)
plt.gca().set_aspect('equal', 'datalim')
#title = 'UMAP projection - color dataset with ', dimensiones ,' dims'
plt.title('UMAP projection of the texture dataset layer 1', fontsize=12)
handles, labels = scatter_plot.legend_elements()
labels = string_labels
plt.legend(handles, labels)
plt.show()
```
